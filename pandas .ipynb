# Merging Joining Concatination

```python
# DataFrace1 : 

employees = pd.DataFrame({
    'employees_id': [1,2,3,4,5],
    'Name': ['john','Anna','Peter','limda','Boss'],
    'department':['HR','It','Finance', 'It','It']
})


#DataFrame 2:
Salaries = pd.DataFrame({
    'employees_id' : [ 1 , 2 , 3 , 6 , 7 ],
    'Salary': [60000, 43000,45000,70000,65000]
})

```

```python
employees
```

```python
Salaries
```

```python
pd.merge(employees, Salaries , on = "employees_id", how = "inner")
```

```python
pd.merge(employees, Salaries , on = "employees_id", how = "outer")
```

```python
pd.merge(employees, Salaries , on = "employees_id", how = "left")
```

```python
pd.merge(employees, Salaries , on = "employees_id", how = "right")
```

# Group by Aggrigation 

```python
data = {
    'Categry' : ['A','B','A','B','A','B'],
    'Store' : ['s1','s2','s2', 's1','s1','s2'],
    'Sale' : [100 , 200 , 300 , 400 , 500 , 600],
    'Quanty': [10 , 20 , 30 , 40 , 50 , 60],
    'Date' : pd.date_range('2026 - 06 - 13', periods = 6)
}

df8 = pd.DataFrame(data)
```

```python
df8
```

```python
df8[["Sale","Quanty"]]
```

```python
# group by categry and calculate the sum of sales
cat = df8.groupby('Categry')
cat

for i , v in cat:
    print(i)
    print(v)
```

```python
cat = df8.groupby('Categry')['Sale'].sum()
cat
```

```python
# Group by store and calculate the sum of sales
cat = df8.groupby('Store')['Sale'].sum()
cat
```
